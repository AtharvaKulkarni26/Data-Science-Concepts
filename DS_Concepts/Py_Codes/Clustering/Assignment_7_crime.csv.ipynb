```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
crime = pd.read_csv('crime_data.csv')
crime.head()
```

```python
crime.info()
```

```python
# Normalization function 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_crime_df = scaler.fit_transform(crime.iloc[:,1:])
```

```python
# Hierarchical Clustering
```

```python
# create dendrogram
dendrogram = sch.dendrogram(sch.linkage(scaled_crime_df, method='single'))
```

```python
# create clusters
hc = AgglomerativeClustering(n_clusters=4, affinity = 'euclidean', linkage = 'single')
```

```python
# save clusters for chart
y_hc = hc.fit_predict(scaled_crime_df)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
```

```python
Clusters
```

```python
crime['h_clusterid'] = Clusters
```

```python
crime
```

```python
crime.groupby('h_clusterid').agg(['mean']).reset_index()
```

```python
# K-Means Clustering
```

```python
from sklearn.cluster import KMeans
```

```python
crime1 = pd.read_csv('crime_data.csv')
```

```python
# Normalization function 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_crime1_df = scaler.fit_transform(crime1.iloc[:,1:])
```

```python
# How to find optimum number of cluster?
# The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_crime1_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm
from sklearn.cluster import KMeans
clusters_new = KMeans(4, random_state=42)
clusters_new.fit(scaled_crime1_df)
clusters_new.labels_
```

```python
#Assign clusters to the data set
crime1['clusterid_new'] = clusters_new.labels_
```

```python
crime1
```

```python
crime1.groupby('clusterid_new').agg(['mean']).reset_index()
```

```python
# DBSCAN
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime2 = pd.read_csv('crime_data.csv')
crime2.head()
```

```python
crime2.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
array=crime2.values
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
X
```

```python
dbscan = DBSCAN(eps=1, min_samples=5)
dbscan.fit(X)
```

```python
#Noisy samples are given the label -1.
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
pd.concat([crime2,cl],axis=1)
```
